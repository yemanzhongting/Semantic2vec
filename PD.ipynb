```python
def merge_dict(x,y):
    for k,v in x.items():
                if k in y.keys():
                    y[k] += v
                else:
                    y[k] = v
    return y
```

```python
def adddict(A,B):
    for key,value in B.items():
        if key in A:
            A[key] += value
        else:
            A[key] = value
    return(dict(sorted(A.items(), key=lambda B:B[1])))
```
